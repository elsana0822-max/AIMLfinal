```python
import numpy as np
from rdkit.Chem import AllChem

fpgen = AllChem.GetRDKitFPGenerator()

fp_total = []
for index, row in df.iterrows():
    smi = row['SMILES']
    mol = Chem.MolFromSmiles(smi)
    fp = fpgen.GetFingerprint(mol)
    fp_array = np.zeros((0,), dtype=int)
    Chem.DataStructs.ConvertToNumpyArray(fp, fp_array)
    fp_total.append(fp_array)
```

```python
# smiles 코드가 단일 구조가 아니라면 '.'이 포함되어 있음.
# 단일 구조가 아닌 물질들만 리스트를 확인하고, 제거해야 될 것과 제거하면 안되는 물질 구분해보기.
# 현재 대부분이 유기 화합물이기 때문에, 유기화합물이 염(salt)가 추가된 경우는, 염을 제거한 smiles 코드를 남겨야함.
# 무기화합물이라면 제거. 유효 성분이 몇개인지 확인이 어려운 경우도 제거.
```

```python
# 'SMILES' 열에서 '.'이 포함된 행만 추출
multi_structures = df[df['SMILES'].str.contains(r'\.', na=False)]
print(multi_structures['SMILES'])
```

```python
def clean_smiles(smiles):
    if '.' not in smiles:
        return smiles # 단일 구조는 그대로 반환
    
    # 1. '.'을 기준으로 조각내기
    fragments = smiles.split('.')
    
    # 2. 각 조각의 길이를 기준으로 내림차순 정렬 (가장 긴 게 앞으로)
    # 보통 유기 화합물 본체가 염보다 SMILES 길이가 깁니다.
    fragments.sort(key=len, reverse=True)
    
    best_fragment = fragments[0]
    
    # 3. 무기 화합물 체크 (탄소 'C' 또는 'c'가 포함되어 있는지 확인)
    if 'C' not in best_fragment.upper():
        return None # 탄소가 없으면 제거 대상
    
    # 4. 유효 성분 판단 (첫 번째와 두 번째 조각의 길이 차이가 별로 안 나면 위험)
    if len(fragments) > 1 and len(best_fragment) < len(fragments[1]) * 1.2:
        # 두 덩어리의 크기가 비슷하면 혼합물일 가능성이 높으므로 제거
        return None
        
    return best_fragment

# 데이터프레임에 적용
df['SMILES_Clean'] = df['SMILES'].apply(clean_smiles)

# 결과 확인: None으로 변한 것들은 제거(dropna)
df_final = df.dropna(subset=['SMILES_Clean'])

print(f"정제 전: {len(df)}개 -> 정제 후: {len(df_final)}개")
```

```python
# 1. 원래 데이터에서 '.'이 포함된 것들만 따로 뽑기
dots_only = df[df['SMILES'].str.contains(r'\.', na=False)].copy()

# 2. 정제 함수 적용 (이전에 만든 clean_smiles 함수 사용)
dots_only['Cleaned'] = dots_only['SMILES'].apply(clean_smiles)

# 3. 제거된 것(None)과 남은 것 비교해서 출력
removed = dots_only[dots_only['Cleaned'].isnull()]
kept = dots_only[dots_only['Cleaned'].notnull()]

print(f"점(.)이 있던 데이터 총 {len(dots_only)}개 중:")
print(f"- 완전히 제거된 물질: {len(removed)}개")
print(f"- 염만 제거하고 남긴 물질: {len(kept)}개")

print("\n--- [제거된 물질들 (무기물 또는 판단불가)] ---")
print(removed['SMILES'].unique())

print("\n--- [남겨진 물질들 (염 제거 후 본체만 남음)] ---")
print(kept[['SMILES', 'Cleaned']].head(10))
```

```python
from rdkit import Chem

def refine_smiles(smiles):
    # 1. 단일 구조면 그대로 반환 (단, 탄소가 없는 무기물은 제외)
    if '.' not in smiles:
        mol = Chem.MolFromSmiles(smiles)
        if mol and any(atom.GetSymbol() == 'C' for atom in mol.GetAtoms()):
            return smiles
        else:
            return None

    # 2. 다중 구조(.) 처리
    fragments = smiles.split('.')
    # 각 조각을 RDKit 객체로 변환
    mols = [Chem.MolFromSmiles(f) for f in fragments]
    # None이 섞여있을 수 있으니 유효한 것만 필터링
    mols = [m for m in mols if m is not None]
    
    if not mols:
        return None

    # 각 조각의 원자 개수를 계산 (중요도 판단 기준)
    mol_sizes = [(m, m.GetNumAtoms(), Chem.MolToSmiles(m)) for m in mols]
    # 원자 개수 많은 순으로 정렬
    mol_sizes.sort(key=lambda x: x[1], reverse=True)

    biggest_mol, biggest_size, biggest_smiles = mol_sizes[0]

    # [조건 1] 무기화합물 제거: 가장 큰 덩어리에 탄소(C)가 없으면 무기물로 간주하고 버림
    has_carbon = any(atom.GetSymbol() == 'C' for atom in biggest_mol.GetAtoms())
    if not has_carbon:
        return None

    # [조건 2] 유효 성분 확인 불가 제거: 
    # 첫 번째로 큰 조각과 두 번째로 큰 조각의 원자 수 차이가 크지 않으면(예: 3개 미만) 제거
    if len(mol_sizes) > 1:
        second_size = mol_sizes[1][1]
        if (biggest_size - second_size) < 3: # 차이가 너무 적으면 판단 불가로 간주
            return None

    # [조건 3] 유기화합물 본체 남기기: 위 조건을 통과했다면 가장 큰 덩어리만 반환 (염 제거 효과)
    return biggest_smiles

# 데이터프레임 적용
df['SMILES_Refined'] = df['SMILES'].apply(refine_smiles)

# 결과 확인 (None인 것들 버리기)
df_cleaned = df.dropna(subset=['SMILES_Refined']).copy()

print(f"전체 데이터: {len(df)}개")
print(f"조건에 맞지 않아 제거된 데이터: {len(df) - len(df_cleaned)}개")
print(f"최종 남은 데이터: {len(df_cleaned)}개")
```

```python
#https://www.rdkit.org/docs/GettingStartedInPython.html#list-of-available-descriptors
#fingerprint 종류는 매우 다양, 종류별로 모두 계산해서 저장해보기
```

```python
from rdkit.Chem import AllChem
import numpy as np

# 핑거프린트를 숫자로 변환하는 보조 함수
def fps_to_array(fp):
    return np.array(fp)

# 결과물을 담을 리스트
morgan_list = []
maccs_list = []
rdk_list = []

for smiles in df_cleaned['SMILES_Refined']:
    mol = Chem.MolFromSmiles(smiles)
    if mol:
        # 1. Morgan Fingerprint (Radius 2는 보통 4개 결합 거리까지 봅니다)
        fp_morgan = AllChem.GetMorganFingerprintAsBitVect(mol, radius=2, nBits=2048)
        morgan_list.append(list(fp_morgan))
        
        # 2. MACCS Keys (166비트 고정)
        fp_maccs = AllChem.GetMACCSKeysFingerprint(mol)
        maccs_list.append(list(fp_maccs))
        
        # 3. RDKit Fingerprint
        fp_rdk = Chem.RDKFingerprint(mol)
        rdk_list.append(list(fp_rdk))

# 데이터프레임으로 만들기
df_morgan = pd.DataFrame(morgan_list).add_prefix('Morgan_')
df_maccs = pd.DataFrame(maccs_list).add_prefix('MACCS_')
df_rdk = pd.DataFrame(rdk_list).add_prefix('RDKit_')

# 기존 데이터와 합치기
df_final = pd.concat([df_cleaned.reset_index(drop=True), df_morgan, df_maccs, df_rdk], axis=1)

print(f"최종 데이터 형태: {df_final.shape}")
```

```python
df_final.to_csv('skin_irritation_with_fingerprints.csv', index=False)
print("저장 완료!")
```
